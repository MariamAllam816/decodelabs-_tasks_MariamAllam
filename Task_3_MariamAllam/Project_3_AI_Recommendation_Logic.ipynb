```python
import pandas as pd
import io
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
csv_data = """
Job_Role,skills_list
Data Scientist,Python SQL Machine Learning Data Analysis Pandas TensorFlow
DevOps Engineer,AWS Docker Kubernetes CICD Linux Python Automation
Backend Developer,Java Python SQL APIs Nodejs Git REST
Cloud Architect,AWS Cloud Computing Networking Security Python Terraform Cloud
Frontend Developer,JavaScript HTML CSS React UIUX Web Design Frontend
"""
df = pd.read_csv(io.StringIO(csv_data))
```

```python
user_profile_str = "Python SQL Machine Learning" # Example input
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['skills_list'])
user_vector = vectorizer.transform([user_profile_str])
```

```python
similarity_scores = cosine_similarity(user_vector, tfidf_matrix).flatten()
df['Match_Score'] = similarity_scores

recommendations = df.sort_values(by='Match_Score', ascending=False).head(3)
```

```python
print("---YOUR top recommendations---")
for index, row in recommendations.iterrows():
    # Convert decimal to percentage safely
    match_percentage = row['Match_Score'] * 100
    print(f"Role: {row['Job_Role']} | Match Confidence: {match_percentage:.2f}%")
```
