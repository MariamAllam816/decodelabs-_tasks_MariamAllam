```python
import re
```

```python
intents = {
    "hi": "Hello! How can I help you today?",
    "hello": "Hi there! How may I assist you?",
    "hey": "Hello! How can I help you today?",
    "how are you": "I'm just a bot, but I'm doing well. How about you?",
    "tell me a joke": "Why don't skeletons fight each other? They don't have the guts!"
}
```

```python
def sanitize_input(raw_input):
    clean_text = raw_input.lower().strip()
    clean_text = re.sub(r'[^\w\s]', '', clean_text)
    return clean_text
```

```python
def process_logic(clean_input):
    fallback = "I'm sorry, I didn't understand that. Could you rephrase?"
    return intents.get(clean_input, fallback)
```

```python
def chat_with_bot():
    print("Hello, I am your rule-based Logic Engine! Type 'exit' to end.")

    # INFINITE CYCLE[cite: 1]
    while True:
        user_input = input("You: ")
        clean_input = sanitize_input(user_input)

        # KILL COMMAND[cite: 1]
        if clean_input in ['exit', 'bye', 'quit']:
            print("Chatbot: Goodbye! Have a great day!")
            break

        # PROCESS & OUTPUT[cite: 1]
        response = process_logic(clean_input)
        print(f"Chatbot: {response}")
```

```python
chat_with_bot()
```
